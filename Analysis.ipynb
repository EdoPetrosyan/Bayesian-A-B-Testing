### Importing our created classes for further analysis of two models to do bayesian A/B testing.

```python
from BanditClass import Bandit, EpsilonGreedy, ThompsonSampling, Visualization, comparison
```

### Defining the rewards for each bandit and the number of trials for the experiment. This setup will be used for both the Epsilon-Greedy and Thompson Sampling strategies.


```python
Bandit_Reward = [1, 2, 3, 4]
NumberOfTrials = 20000
```

###  Running Epsilon-Greedy and Thompson Sampling experiments

```python
(cumulative_avg_reward_1, cumulative_total_reward_1, cumulative_regrets_1, greedy_bandits,
 bandit_choices_1, rewards_1, suboptimal_pull_count) = EpsilonGreedy(Bandit).experiment(Bandit_Reward, NumberOfTrials)
Visualization().plot_cumulative_reward(NumberOfTrials, cumulative_avg_reward_1, greedy_bandits, 'Epsilon Greedy')
```

What we observe here, is that the Eps. Greedy model was stuck on suboptimal bandit and could not reach to optimal bandit, which is of course very inefficient, though this demonstrates greatly drawback of this model.

### Visualizing Thompson Sampling Results

```python
(cumulative_avg_reward_2, cumulative_total_reward_2, 
     cumulative_regrets_2, thompson_bandits, 
     bandit_choices_2, rewards_2) = ThompsonSampling(Bandit).experiment(Bandit_Reward, NumberOfTrials)
```

As we see, after many trials, we can truly see, that the optimum bandit is the one having rate 4 and its distribution has mean 4 and very small variance.

```python
Visualization().plot_cumulative_reward(NumberOfTrials, cumulative_avg_reward_2, thompson_bandits, 'Thompson Sampling')
```

### Comparing Performance of Both Algorithms.

```python
Visualization().compare_algorithms(cumulative_total_reward_2, cumulative_total_reward_1, cumulative_regrets_2, cumulative_regrets_1)
```

```python
comparison(NumberOfTrials, cumulative_avg_reward_1, cumulative_avg_reward_2, 
           cumulative_total_reward_1, cumulative_total_reward_2, 
           cumulative_regrets_1, cumulative_regrets_2, 
           thompson_bandits)
```

As we see, Thompson Sampling algorithm did more accurate job of finding the optimal bandit.

### Generating Reports

```python
EpsilonGreedy(Bandit).report(NumberOfTrials, greedy_bandits, 
                             bandit_choices_1, rewards_1, 
                             cumulative_regrets_1, suboptimal_pull_count, 
                             'EpsilonGreedy')
```

```python
ThompsonSampling(Bandit).report(NumberOfTrials, thompson_bandits, 
                             bandit_choices_2, rewards_2, 
                             cumulative_regrets_2, 
                             algorithm='ThompsonSampling')
```
